## II. Feature extraction

### II.a The iterative process:
I start with a few hand crafted features I can think of and follow these iterative steps below to find more features. At the end of each random forest prediction, I look at the misclassified samples and try to study from the text what features could help improve the model: </n>

Random forest -> Study misclassified -> Feature extraction -> Random forest ->...</n>

Fit models using: </n>
```python
from sklearn.model_selection import train_test_split
X = df[list of features]
y = df['class_label']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y)
```

## III. Finalise model

### III.a Random forest

Random search CV is conducted on the training data to find the best hyperparameters.

````python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import make_scorer, f1_score
scoring = make_scorer(f1_score, average='micro')

X = df.drop(['class_label', 'class_id', 'url', 'text', 'no_ents_text', 'lem_text'], axis=1)
# X = df[['len', 'org_count', 'place_count', 'time_count', 'person_count', 'num_count', 'ents_rep', 'vocab', 'ppo_rep', 'verb_present', 'verb_past', 'verb', 'modal', 'ad', 'prob']]
y = df['class_label']
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, stratify=y, random_state=2020)
rf = RandomForestClassifier()
parameter_space = {
                'n_estimators': [800, 1000, 1500],
                'max_depth': [5, 8, None],
                'min_samples_split': [2, 5],
                'min_samples_leaf': [1, 2, 5],
                'bootstrap': [True]
            }

rf_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=parameter_space,
    scoring=scoring,
    n_iter=10,
    cv=10,
    verbose=0,
    random_state=42,
    n_jobs=-1
)

rf_search.fit(X_train, y_train)
print('Best parameters are:', rf_search.best_params_)
print()
print('F1 score:', rf_search.best_score_)
print()

y_test_pred = rf_search.best_estimator_.predict(X_test)
print('Test result:', classification_report(y_test, y_test_pred))
````

### III.b CatBoost

````python
import catboost as cb
parameter_space = {
    'depth': [4, 7],
    'learning_rate': [0.15, 0.3, 0.8],
    'l2_leaf_reg': [4, 9, 15],
    'iterations': [500]
}
cb_clf = cb.CatBoostClassifier()
cb_search = RandomizedSearchCV(
    estimator=cb_clf,
    param_distributions=parameter_space,
    scoring=scoring,
    n_iter=10,
    cv=10,
    verbose=0,
    random_state=42,
    n_jobs=-1
)

cb_search.fit(X_train, y_train)
print('Best parameters are:', cb_search.best_params_)
print()
print('F1 score:', cb_search.best_score_)
print()

y_test_pred = cb_search.best_estimator_.predict(X_test)
print('Test result:', classification_report(y_test, y_test_pred, target_names=cls_names))
````

There might be better hyperparameters for CatBoost what produce better result. However for the limited time for training, random forest is sufficient and very quick to run too.
